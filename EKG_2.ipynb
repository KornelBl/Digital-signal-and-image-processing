```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Punkt 1.

```python
f = 50 #Hz
N = 65536
T = 1 #s
fs = 65536/5 #Hz
n = np.arange(N)
```

```python
phasor = (n * 2 * np.pi * f / fs) % (2 * np.pi)
plt.plot(n, phasor)
plt.xlabel('nr próbki')
plt.ylabel('kąt fazowy [rad]')
plt.title('Faza')
plt.show()
```

```python
sinus = np.sin(phasor)
plt.plot(n, sinus)
plt.xlabel('nr próbki')
plt.ylabel('amplituda')
plt.title('Sygnał sinusoidalny')
plt.show()
```

Punkt 2.

```python
spectrum = np.fft.fft(sinus)
xs = np.linspace(0,fs/2,N//2)
plt.plot(xs, np.abs(spectrum[:(N//2)]))
plt.xlabel('nr próbki')
plt.ylabel('amplituda')
plt.title('Sygnał sinusoidalny')
plt.show()
```

Punkt 3.

```python
f1 = 50 #Hz
f2 = 60 #Hz
N = 65536
fs = 150 #Hz
n = np.arange(N)
```

```python
phasor1 = (n * 2 * np.pi * f1 / fs) % (2 * np.pi)
phasor2 = (n * 2 * np.pi * f2 / fs) % (2 * np.pi)
sinus = np.sin(phasor1) + np.sin(phasor2)
plt.plot(n, sinus)
plt.xlabel('nr próbki')
plt.ylabel('amplituda')
plt.title('Sygnał sinusoidalny')
plt.show()
```

```python
spectrum = np.fft.fft(sinus)
xs = np.linspace(0,fs/2,N//2)
plt.plot(xs, np.abs(spectrum[:(N//2)]))
plt.xlabel('nr próbki')
plt.ylabel('amplituda')
plt.title('Sygnał sinusoidalny')
plt.show()
```

Punkt 4.

```python
fig = plt.figure()

fig.plot(xs, np.abs(spectrum[:(N//2)]))
fig.xlabel('nr próbki')
fig.ylabel('amplituda')
fig.title('Sygnał sinusoidalny')
fig.show()
```
