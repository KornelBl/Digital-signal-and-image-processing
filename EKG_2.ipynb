```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,9)
```

```python
def plot_spectrum(fft_spectrum :np.array,f_s:int) -> None:
    n_s = len(fft_spectrum)
    x = np.linspace(0,f_s/2,(n_s//2)+1)
    plt.bar(x, np.abs(fft_spectrum[:(n_s//2)+1]))
    plt.xlabel('Hz')
    plt.ylabel('widmo')
    plt.title('Widmo sygnału')
    plt.show()

def plot_signal(signal :np.array,n: int = None,title = "Sygnał sinusoidalny") ->None:
    if n is None:
        n = len(signal)
    n_axis = np.arange(n)
    plt.plot(n_axis, signal[:n])
    plt.xlabel('nr próbki')
    plt.ylabel('amplituda')
    plt.title(title)
    plt.show()     
```

Punkt 1.

```python
f = 50#Hz
N = 65536
fs = 120 #Hz
n = np.arange(N)
```

```python
phasor = (n * 2 * np.pi * f / fs) % (2 * np.pi)
sinus = np.sin(phasor)
plot_signal(sinus,200)
```

Punkt 2.

```python
spectrum = np.fft.fft(sinus,120)
plot_spectrum(spectrum,fs)
```

```python
f1 = 50 #Hz
f2 = 60 #Hz
N = 65536
fs = 500  #Hz
```

```python
spectrum = np.fft.fft(sinus,fs)
plot_spectrum(spectrum,fs)
```
